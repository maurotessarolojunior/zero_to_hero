# Alter upper and lower case in string

```python
def alter(**kwargs):
    if 'alt' in kwargs:
        alt = kwargs['alt']
        n = len(alt)
        f=''
        for x in range(n):
            if (x%2) ==0:
                f+=alt[x].upper()
            else:
                f+=alt[x]  
        print(f)
```

```python
alter(alt='mauro')
```

# Enumerate Function

```python
index_count = 0

word = 'abcde'

for letter in word:
    print(word[index_count])
    index_count+=1
```

```python
word = 'abcde'

for item in enumerate(word):
    print(item)
```

```python
word = 'abcde'        # From Tupple Unpacking

for index, letter in enumerate(word):
    print(letter)
    print(index)
    print('\n')
```

So what enumerate returns is the **index** and the **object** at that index

```python
# Old Way of doing things:
```

```python
jupfile = open('jupfile.txt')
```

```python
# New Way:
```

```python
%%writefile jupfile.txt
Hi, My name is MJ
My second line
Is my third try
```

with open('jupfile.txt') as jupfile:
    # Whatever code I am writing here it is going to use jupfile as a variable of jupfile.txt

```python
# New way here:
```

```python
with open('jupfile.txt') as jf:
    A = jf.read()
```

```python
A
```

```python
# And now we no longer need to close the file
```

```python
# Why? Because it is acting upon the variable (jupfile) and not upon the original .txt file
```

```python
with open('jupfile.txt',mode='r') as jp:
    A = jp.read()
```

```python
A
```

```python
print(A)
```

```python
print(open('jupfile.txt',mode='r').read()) # = print(A)
```

```python
with open('jupfile.txt',mode='r') as jp:
    print(jp.readlines())
```

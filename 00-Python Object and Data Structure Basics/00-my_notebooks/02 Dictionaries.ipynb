When to use it?

When you don't have the need to know the index of a list and just want to retrieve information

And neither reverse or append?<br>
You CAN add or change an item by directly informing the 'string ref':item <br>
Or you can remove it using .pop('string ref')<br>
But you can't reverse, because you can't index or slice<br>

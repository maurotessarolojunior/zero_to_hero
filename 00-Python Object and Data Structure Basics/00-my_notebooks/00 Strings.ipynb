<BR><BR><BR>***Print Formatting with Strings***

The first obvious suggestion to insert a variable into a string is:

```python
name = "Arthur"
```

```python
print ("Hello " + name)
```

<br><br>But there are ***two main ways*** *to FORMAT and INSERT TEXT into a string*


**1)**

.format ( ) method

' string  { } more string { } ' .format ( 'something1' , 'something2' )

e.g.

```python
print('Put this word {}'.format('HERE'))       #OPTION 1
```

```python
print ('My {} {} {}'.format('mauro','is','name'))
```

```python
print ('My {2} {1} {0}'.format('mauro','is','name')) #As variáveis possuem indexes
```

```python
print ('faça o {c} da {f} {t}'.format(c='contrato',f='família',t='teixeira')) #Aqui foi atribuído keywords aos Indexes
```

```python
print ('faça o {c} da {c} {c}'.format(c='contrato',f='família',t='teixeira'))
```

<br>

Float formatting

" { value : width.precision f } "

Width only ads zeros to the right that don't even appear, it just changes how long you want the string to be

Whereas precision is the decimal numbers you want to appear

```python
result = 100/777
```

```python
result
```

```python
print ('the result was {}'.format(result))       
```

```python
print ('the result was {r:.3f}'.format(r=result)) #para que possa usar a formatação precisei atribuir keyword ao resultado
```


<br>**2)**                                               

f-strings (formatted string literals)

```python
print (f'the result was {result}')                     #OPTION 2
```

```python
print (f'the result was {result:1.3f}')
```

```python
print (f'my name is {'mauro'}')
```

```python
name= "Mauro"
age= 32
```

```python
print (f'{name} is {age} years old')
```
